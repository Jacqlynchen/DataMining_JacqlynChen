```python
import pandas as pd

# Download the CSV file from the URL
# Load the dataset and display the first 5 rows
url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url, sep=',')

df.head()
```

```python
# Show the dataset info
df.info()
```

Based on the dataset info, we can see that the dataset contains 3553 entries and 27 columns. However, several columns show a noticeable number of non-null values that are less than the total entries, indicating the presence of missing data that will need to be handled before further analysis.

```python
# Show the descriptive statistics
df.describe()
```

Key Findings from the dataset
1. Price
  - The price values have a very wide range, from extremely low to very high.
  - The mean price is higher than the median → this indicates a right-skewed distribution, meaning most houses are relatively affordable, but a few luxury properties push the average upward.
2. Land Area & Building Size
  - Both land and building sizes show large variation, ranging from very small to several thousand square meters.
  - This suggests the presence of outliers, such as villas or commercial properties.
  - The interquartile range (Q1–Q3) provides a clearer picture of the typical house size in Jabodetabek.
3. Categorical Variables (city, property_type, etc)
  - Certain cities or districts appear more frequently than others → indicating a concentration of listings in specific areas.
  - The dominant property type is likely residential houses, while other types (apartments or shop houses) are less common.
4. Rooms (bedrooms, bathrooms, carports, garages)

```python
# Display unique values for each column
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
df = df.drop(['url', 'ads_id', 'title', 'lat', 'long', 'facilities'], axis=1, errors='ignore')
df.head()
```

After checking the unique values for each column, some variables were found to contain either irrelevant information, inconsistencies, or outliers. These columns or values were then dropped to improve the dataset quality and ensure that only meaningful features remain for further analysis.

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Fill missing values with the median for numerical columns
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())
df['land_size_m2'] = df['land_size_m2'].fillna(df['land_size_m2'].median())
df['building_size_m2'] = df['building_size_m2'].fillna(df['building_size_m2'].median())
df['floors'] = df['floors'].fillna(df['floors'].median())
df['year_built'] = df['year_built'].fillna(df['year_built'].median())
```

Using the median to fill missing values in numerical columns to ensure data completeness, maintain dataset quality, and reduce the impact of outliers on your analysis.

```python
print(df[['bedrooms','bathrooms','land_size_m2','building_size_m2','floors','year_built']].isnull().sum())
```

```python
# Fill the missing values in building_age using the difference between 2022 and year_built
df['building_age'] = df['building_age'].fillna(2022 - df['year_built'])
```

Filled missing values in building_age by calculating them directly from year_built, ensuring accurate and consistent property age information across the dataset.

```python
cat_cols = ['certificate', 'property_condition', 'furnishing', 'building_orientation']
for col in cat_cols:
  if col in df:
    df[col] = df[col].fillna('Unkonown', inplace=True)
```

This ensures no missing categories remain while still preserving the information that the original data was not provided.

```python
# Binning price_in_rp into categorical ranges
price_bins = [0, 1e9, 3e9, 1e10, float('inf')]
price_labels = ['Low (<1B)', 'Medium (1B-3B)', 'High (3B-10B)', 'Luxury (>10B)']

df['price_bin'] = pd.cut(
    df['price_in_rp'],
    bins=price_bins,
    labels=price_labels
)

# Bar plot for price bins
import matplotlib.pyplot as plt
import seaborn as sns

price_counts = df['price_bin'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=price_counts.index, y=price_counts.values, palette='viridis')
plt.title('Distribution of House Prices')
plt.xlabel('Price Range')
plt.ylabel('Number of Houses')
plt.xticks(rotation=45)
plt.show()
```

From the bar chart “Distribution of House Prices”:

1. The largest categories are Medium (19–30) and Low (≤18).

    - Each of these categories has more than 1,200 houses.

    - This indicates that most houses fall within the affordable or mid-range price levels.

2. The High (30–100)
    - category is smaller compared to the first two, with only around 700–800 houses.

    - This suggests that the availability or demand for houses in the upper-mid segment is lower.

3. The Luxury (>100)
    - category is the smallest, with only about 300–400 houses.

    - This shows that very high-priced houses are rarely available or less in demand.

4. The pattern is clear
    - the higher the house price, the fewer houses are available or owned by the public.

The housing market is dominated by the low and medium price range segments, indicating that the majority of buyers or developers focus on affordable housing. Meanwhile, the high and luxury segments tend to be niche markets with more limited demand.

```python
# Binning land_size_m2 into categorical ranges
land_bins = [0, 100, 250, 1000, float('inf')]
land_labels = ['Small (<100 m2)', 'Medium (100–250 m2)', 'Large (250–1000 m2)', 'Extra Large (>1000 m2)']

df['land_size_bin'] = pd.cut(
    df['land_size_m2'],
    bins=land_bins,
    labels=land_labels
)

# Bar plot for land size bins
land_counts = df['land_size_bin'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=land_counts.index, y=land_counts.values, palette='magma')
plt.title('Distribution of Land Size')
plt.xlabel('Land Size Range (m²)')
plt.ylabel('Number of Houses')
plt.xticks(rotation=45)
plt.show()
```

The housing market is dominated by houses with small to medium land sizes, while large and extra-large plots only serve a limited segment. This reflects the greater demand for practical and affordable housing compared to properties with expansive land areas.
